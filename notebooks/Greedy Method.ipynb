```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pickle

expmt = '501_0.1_0.2_uniform_0.5_0.5_uniform_0.01_False_50_5'
method = 'Weighted'
dataset_name = 'abalone'

weighted = pd.read_csv('./results/' + dataset_name + '/' + method + '/' + expmt + '/results_df')
weighted['rmse'] = np.sqrt(weighted['mse'])
afasmc = pd.read_csv('./results/' + dataset_name + '/AFASMC/' + expmt + '/results_df')
# rand = pd.read_csv('./results/abalone/Rand/'  + expmt + '/results_df')
# abalone = pd.concat([weighted, afasmc, rand])
abalone = pd.concat([weighted, afasmc])

sns.lineplot(x='n_available', y='mse', hue='acq_model', data=abalone, marker='o')
```

```python
expmts = ['501_0.1_0.2_uniform_0.5_0.5_uniform_0.0_False_50_5',
        '501_0.1_0.2_uniform_0.5_0.5_uniform_0.01_False_50_5',
        '501_0.1_0.2_uniform_0.5_0.5_uniform_0.1_False_50_5',
        '501_0.1_0.2_uniform_0.5_0.5_uniform_0.25_False_50_5',
        '501_0.1_0.2_uniform_0.5_0.5_uniform_0.5_False_50_5',
        '501_0.1_0.2_uniform_0.5_0.5_uniform_1_False_50_5']
l_vals = [0, .01, .1, .25, .5, 1]

method = 'Weighted'
dataset_name = 'abalone'
all_results = []
for expmt, l_val in zip(expmts, l_vals):
    weighted = pd.read_csv('./results/' + dataset_name + '/' + method + '/' + expmt + '/results_df')
    weighted['rmse'] = np.sqrt(weighted['mse'])
    weighted['l_val'] = l_val
    all_results.append(weighted)
all_results = pd.concat(all_results)
sns.lineplot(x='n_available', y='cost_var', hue='l_val', data=all_results, marker='o')
plt.xlabel("User Privacy Cost")
plt.ylabel("RMSE")
plt.title("Inclusion of Privacy Cost: Effect on Privacy Cost")
```
