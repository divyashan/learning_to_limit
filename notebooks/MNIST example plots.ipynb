```python
%matplotlib inline

import sklearn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import fetch_openml 

mnist = fetch_openml('mnist_784', version=1, return_X_y=True)
```

```python
X, y = mnist
y = np.array([int(i) for i in y])
for i in range(9):
    im_idx = np.argwhere(y == i+1)[0]
    plottable_image = np.reshape(X[im_idx], (28, 28))
    plt.subplot(3, 3, i + 1)
    plt.imshow(plottable_image, cmap='gray_r')
    plt.xticks([])
    plt.yticks([])

plt.plot()
plt.tight_layout()
```

```python
from skimage.transform import rotate
im_idx = np.argwhere(y == 4)[0]
plottable_image = np.reshape(X[im_idx], (28, 28))
plt.imshow(rotate(plottable_image, 30), cmap='gray_r')
plt.xticks([])
plt.yticks([])
```
